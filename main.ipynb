# World Happiness Report Analysis(2019)

## Data Overview

loading Dataset in Dataframe

```python
import pandas as pd

# Loading the dataset
df = pd.read_csv('dataset/2019.csv')

print(df.head())
```

## Data Cleaning
Removing Duplicates values & Replacing Null values

```python
# Data cleaning
df = df.drop_duplicates()  # Removing duplicates
df = df.fillna(0)  # Replace missing values with 0

print(df.head())
```

## Histogram of Happiness Score

```python
import os
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df['Score'], bins=20, color='skyblue', edgecolor='black')
plt.title(f'Distribution of Happiness Scores (2019)')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

## Histogram of GDP per capita

```python
import os
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df['GDP per capita'], bins=20, color='skyblue', edgecolor='black')
plt.title(f'Distribution of GDP per capita (2019)')
plt.xlabel('GDP per capita')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()

```

## Heatmap of correlation (happiess factors)

```python
import os
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_df.corr()    
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title(f'Correlation Heatmap of Happiness Factors (2019)')    
plt.show()
```

## Scatter Plot of GDP per capita vs Happiness Score

```python
import os
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
x_col = 'GDP per capita'
y_col = 'Score'

# Scatter plot
plt.scatter(df[x_col], df[y_col], color='purple', alpha=0.6)
plt.title(f'GDP per Capita vs. Happiness Score (2019)')
plt.xlabel('Economy (GDP per Capita)')
plt.ylabel('Happiness Score')
plt.grid()
plt.show()
```
